# Removing rows with missing values

```python
raw_data_count = recent_grads.loc[:]
recent_grads = raw_data_count.dropna()
recent_grads.loc[:]
```

# Generating a few scatter plots

```python
ax1 = recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter")
ax1.set_title("Sample Size vs Median - Scatter Plot")
```

```python
ax2 = recent_grads.plot(x = "Sample_size", y = "Unemployment_rate", kind = "scatter")
ax2.set_title("Sample Size vs Unemployment_rate - Scatter Plot")
```

```python
ax3 = recent_grads.plot(x = "Full_time", y = "Median", kind = "scatter")
ax3.set_title("Full_time vs Median - Scatter Plot")
```

```python
ax4 = recent_grads.plot(x = "ShareWomen", y = "Unemployment_rate", kind = "scatter")
ax4.set_title("ShareWomen vs Unemployment_rate - Scatter Plot")
```

```python
ax5 = recent_grads.plot(x = "Men", y = "Median", kind = "scatter")
ax5.set_title("Men vs Median - Scatter Plot")
```

```python
ax6 = recent_grads.plot(x = "Women", y = "Median", kind = "scatter")
ax6.set_title("Women vs Median - Scatter Plot")
```

# Plotting Bar Graphs

Plotting percentage of women (defined by variable SharedWomen) for 10 highest and lowest paying majors

```python
recent_grads[:10].plot(x = "Major", y = "ShareWomen", kind = "bar", legend = False)
recent_grads[163:].plot(x = "Major",y = "ShareWomen", kind = "bar", legend = False)
```

Plotting percentage of Unemployment rate for 10 highest and lowest paying majors

```python
recent_grads[:10].plot(x = "Major", y = "Unemployment_rate", kind = "bar", legend = False)
recent_grads[163:].plot(x = "Major",y = "Unemployment_rate", kind = "bar", legend = False)
```

# Key observations
- The median salary for most graduates hovers in the range of 30k - 50k.
- In the 10 highest paid majors, women form a high percentage in the field of Astronomy and Astrophysics while for 10 lowest paid majors, the percentage of women is highest for Early childhood education.
- Nuclear engineering appears to have the highest unemployment rate among high paying majors and Clinical Psychology has the highest unemployment rate among low paying majors.
